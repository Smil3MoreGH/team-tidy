```python
#imports 
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(dpi=1200)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
#import the dataset with pandas
dataFrame = pd.read_excel('./salary/salary_data_states.xlsx')
```

```python
#Did not drop columns, should I have dropped them or is it okay that I kept them? 
#When looking at data subgroups do we only care about data analysts or all jobs that deal with data inclusding software engineer?
#im analyzing software engineer, business analyst, data analyst 

dataFrame['PAID_WAGE_PER_YEAR'] = dataFrame['PAID_WAGE_PER_YEAR'].fillna(0)
dataJobs = dataFrame[(dataFrame["JOB_TITLE_SUBGROUP"] == "software engineer") | (dataFrame["JOB_TITLE_SUBGROUP"] == "business analyst") | (dataFrame["JOB_TITLE_SUBGROUP"] == "data analyst")]
dataJobs
```

```python
dataJobs.groupby("JOB_TITLE_SUBGROUP")["PAID_WAGE_PER_YEAR"].mean()
#data shows that software engineers make the most on average, business analysts rank in second place, with data analysis making the least of the three.
```

```python
# highest salary paid by company 
print("Software Engineer")
print(dataFrame[(dataFrame["JOB_TITLE_SUBGROUP"] == "software engineer")].groupby("EMPLOYER_NAME")["PAID_WAGE_PER_YEAR"].max().sort_values(ascending=False))

print("\nBusiness Analyst")
print(dataFrame[(dataFrame["JOB_TITLE_SUBGROUP"] == "business analyst")].groupby("EMPLOYER_NAME")["PAID_WAGE_PER_YEAR"].max().sort_values(ascending=False))

print("\nData Analyst")
print(dataFrame[(dataFrame["JOB_TITLE_SUBGROUP"] == "data analyst")].groupby("EMPLOYER_NAME")["PAID_WAGE_PER_YEAR"].max().sort_values(ascending=False))
```

```python
#how salary is affected by state 
#yes data subgroup is salary is affected by work state and location
print("Software Engineer")
print(dataFrame[(dataFrame["JOB_TITLE_SUBGROUP"] == "software engineer")].groupby("WORK_STATE")["PAID_WAGE_PER_YEAR"].mean().sort_values(ascending=False))

print("\nBusiness Analyst")
print(dataFrame[(dataFrame["JOB_TITLE_SUBGROUP"] == "business analyst")].groupby("WORK_STATE")["PAID_WAGE_PER_YEAR"].mean().sort_values(ascending=False))

print("\nData Analyst")
print(dataFrame[(dataFrame["JOB_TITLE_SUBGROUP"] == "data analyst")].groupby("WORK_STATE")["PAID_WAGE_PER_YEAR"].mean().sort_values(ascending=False))
```
